## Esercizio 2. Una successione che si “blocca”


Pierino ha inventato uno strano gioco: partendo da un numero naturale
positivo decide che se il numero è pari si dimezza mentre se è dispari si
triplica. Si accorge subito però che, se ripete più volte il procedimento per
creare una successione a partire da un numero, una volta che giunge su un
numero dispari la successione che genera cresce all’infinito. Per rendere la
cosa un po’ più interessante, decide allora di aggiungere 1 al triplo dei dispari.
Le successioni che dunque può creare a partire dai differenti numeri naturali
sono tali che

$$ a_{n+1} = \left \{ \begin{array}{rl}
&\frac{a_n}{2}  &\text{se} \, a_n \, \text{è pari}\\
&3 a_n +1  &\text{altrimenti}
\end{array}
\right. $$

Anche in questo caso, però, Pierino rimane deluso. Infatti, partendo da
un qualunque numero compreso tra 1 e 10, si accorge che la successione si
“incastra” nel ciclo infinito 1, 4, 2, 1, 4, . . ..


### Punto 1 
Implementa in python una funzione chiamata clz che dato an restituisca
$a_{n+1}$ secondo la regola vista poco sopra

```python
# Definiamo una funzione clz(a)
def clz(a):

# Utilizziamo la divisione modulo 2 (operatore modulo (%) che viene utilizzato nella 
# divisione tra numeri interi per conoscere se c'è resto o meno) per selezionare gli
# interi pari con un if statement. Effettuiamo una divisione di interi (//) se la 
# condizione è soddisfatta, in modo da lavorare solo con interi ed evitare successive
# trasformazioni in float.

    if a%2 == 0: 
        a = a//2  
        
    else:
        a = 3*a + 1
        
    return a

#Output di prova
print(clz(5))
print(clz(4))
print(clz(10))
```

2. Diremo che una successione termina se arriva ad 1 (e quindi poi entra nel ciclo infinito con 4, 2 ed 1). Quante successioni terminano tra
quelle per cui $a_1$ è scelto tra i primi 100 numeri naturali positivi?
Notate che per un computer è impossibile affermare con sicurezza
che una successione non termini. Tuttavia, quello che potete fare è
verificare quali successioni terminino entro un numero ragionevole di
passi e, per le rimanenti, verificare a mano (cioè senza un computer)
se stiano divergendo, se siano incastrate in un loop diverso da (4, 2, 1)
o se comunque termineranno ad 1, solo in un numero maggiore di
passi. Ovviamente, per queste successioni, potete usare il computer.
3. Data una successione che termina, la sua “lunghezza” è il minimo n tale per cui $a_n = 1$. Quanto vale la lunghezza della successione più
lunga tra quelle per cui $a_1$ è scelto tra i primi 100 numeri naturali
positivi?

```python
#Inizializzazione variabili
max = 0
N = 101
passi = 100 #Numero passi di iterazione settato come variabile globale modificabile

for i in range(1, N): #ciclo sui primi 100 numeri naturali positivi
    
    j = i 
   
    
    for k in range(passi): #ciclo sui passi di iterazione della successione
        
        #Implementazione della successione
        j = clz(j)
                    
        if j == 1: #criterio di uscita quando la successione "termina"
            print("a_i = {0} questa successione si blocca al {1} ciclo".format(i, k+1))
            if k + 1 > max:
                max = k + 1 #valore del massimo
                i_max = i #punto di massimo
            break
            
        if k == passi-1: #avviso quando la successione "non termina" entro tot cicli
            print("a_i = {0} questa successione non si blocca entro {1} cicli".format(i, passi))
            
print("lunghezza massima = " + str(max))
print("Ed è associata al numero " + str(i_max))
```

Nota: Aumentando il numero di passi, alcune successioni che non terminavano, ora terminano. 
Ad esempio la successione con $a_n=27$ non termina con 100 passi ma termina con 1000 passi, come si può vedere qui di seguito.

```python
#Inizializzazione variabili
max = 0
N = 101
passi = 1000 #Numero passi di iterazione settato come variabile globale modificabile

for i in range(1, N): #ciclo sui primi 100 numeri naturali positivi
    
    j = i 
   
    
    for k in range(passi): #ciclo sui passi di iterazione della successione
        
        #Implementazione della successione
        j = clz(j)
                    
        if j == 1: #criterio di uscita quando la successione "termina"
            print("a_i = {0} questa successione si blocca al {1} ciclo".format(i, k+1))
            if k + 1 > max:
                max = k + 1 #valore del massimo
                i_max = i #punto di massimo
            break
            
        if k == passi-1: #avviso quando la successione "non termina" entro tot cicli
            print("a_i = {0} questa successione non si blocca entro {1} cicli".format(i, passi))
            
print("lunghezza massima = " + str(max))
print("Ed è associata al numero " + str(i_max))
```
