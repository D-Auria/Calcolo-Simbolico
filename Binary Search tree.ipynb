```python
class B_Node:
    
    
    def __init__(self, val):
        self.valore = val
        self.sinistra = None   # "None are tree leaves"
        self.destra = None
        
        
    def __repr__(self):
        
        x = "nessun figlio"

        if self.sinistra is None and self.destra is None:
            return '(Nodo {}, ({}))'.format( self.valore, x)
        
        elif self.sinistra is not None and self.destra is None :
            return '(Nodo {}, (<-- {},--> {}))'.format( self.valore, self.sinistra ,x)
        
        elif self.sinistra is  None and self.destra is not None :
            return '(Nodo {}, (<-- {},--> {}))'.format(self.valore, x, self.destra)
        else :
            return '(Nodo {}, (<-- {},--> {}))'.format( self.valore, self.sinistra, self.destra)
        
        
    def inserisci_valore(self, nuovo_val):
        #print(nuovo_val)
        if nuovo_val < self.valore:
            if self.sinistra is None:
                self.sinistra = B_Node(nuovo_val)
            else:
                self.sinistra.inserisci_valore(nuovo_val)
                
        elif nuovo_val > self.valore:
            if self.destra is None:
                self.destra = B_Node(nuovo_val)
            else:
                self.destra.inserisci_valore(nuovo_val)
        else:
            print("Valore uguale al nodo")
    
    
    def search(self, obj):
        
        if obj == self.valore:
            return(True)
        
        elif obj < self.valore:
            if self.sinistra is None:
                return(False)
            else:
                return self.sinistra.search(obj)
            
        elif obj > self.valore:
            if self.destra is None:
                return(False)
            else:
                return self.destra.search(obj)
           
        
    def altezza_left(self):   
        
        h_sin = 0
         
        #if self is None:
         #   return 0
        if self.sinistra is None:
            return 0
        else:
            if self.sinistra is not None:
                #h_sin = self.sinistra.altezza_left() + 1
                h_sin = self.sinistra.altezza() + 1
                return h_sin
                


    def altezza_right(self):    
        
        h_des = 0
         
        #if self is None:
            #return 0
        if self.destra is None:
            return 0
        else:
            if self.destra is not None:
                #h_des = self.destra.altezza_right() + 1
                h_des = self.destra.altezza() + 1
                return h_des
                
        
    def altezza(self):    
        
        h_sin = self.altezza_left()
        h_des = self.altezza_right()

        if h_des> h_sin:  #cerco quale tra i rami di destra o di sinistra sia più lungo
            return h_des

        else:
            return h_sin
        
    def sbilanciamento(self):
        sb = self.destra.altezza_left() - self.altezza_right()
        return sb
    
    def minimo_nodo(self):
        if self.sinistra is None:

            return self.valore
        else:
            return self.sinistra.minimo_nodo()
        
    def massimo_nodo(self):
        if self.destra is None:
            return self.valore
        else:
            return self.destra.massimo_nodo()
        
    def estrai_valore(self):
        if self.sinistra is None:
            return self.valore
        elif self.destra is None:
            return self.valore
        
   
    def in_lista(self, list):
        list.append(self)                 #metto il nodo nella lista
        if self.sinistra is not None:
            self.sinistra.in_lista(list)
        else:
            return list
 
```

```python
class BinaryTree:
    def __init__(self):
        self.root_node = None
        
    def __repr__(self):
        if self.root_node is None:
            return 'Albero vuoto'
        else:
            print("Albero: ")
            return self.root_node.__repr__()
    
    def InsertValue(self, value):
        if self.root_node is None:
            self.root_node = B_Node(value) 
            #print("ciao")
        else:
            self.root_node.inserisci_valore(value) #attenzione, qui sto usando l'attributo di inserimento valore di B_Node
            
    def __contains__(self, value):
        if self.root_node is None:
            return(False)
        else:
            return self.root_node.search(value)
        
    def altezza_left_tree(self):
        if self.root_node is None:
            return -1
        else:
            return "Altezza sinistra dell'albero: {}".format(self.root_node.altezza_left())
        
    def altezza_right_tree(self):
        if self.root_node is None:
            return -1
        else:
            return "Altezza destra dell'albero: {}".format(self.root_node.altezza_right())
        
    def altezza_tree(self):
        if self.root_node is None:
            a = -1
            return "Altezza dell'albero: {}".format(a)  #per definizione, l'altezza di un albero vuoto, senza nodi è -1
        else:
            return "Altezza dell'albero: {}".format(self.root_node.altezza())
    
    def minimo_tree(self):
        if self.root_node is None:
            return "nessun valore"
        else:
            return "Valore minimo dell'albero: {}".format(self.root_node.minimo_nodo())
        
    def massimo_tree(self):
        if self.root_node is None:
            return "nessun valore"
        else:
            return "Valore massimo dell'albero: {}".format(self.root_node.massimo_nodo())
        
    def sbilanciamento_tree(self):
        return "Sbilanciamento albero (sinistra - destra): {}".format(self.root_node.sbilanciamento())
    
    #def lista_sinistra_tree(self, lista_tree):
     #   if self.root_node is None:
      #      return []
       # else:
        #    return self.root_node.lista_sinistra(lista_tree)
        
    def __iter__(self):
        return BSTIterator(self)
    
```

Naturalmente se ordiniamo l'array di numeri random prima di inserire i suoi valori nell'albero, avremo un albero completamente sbilanciato in uno dei due rami e l'albero "degenera" in una lista di nodi solo destri o sinistri.
Se lasciamo l'array random non ordinato (come nel primo esempio), avremo un albero "in media poco sbilanciato".

```python
class BSTIterator:
    
    def __init__(self, albero):
        self.tree = albero
        self.list = []
        root = self.tree.root_node      
        
        if root is not None:
            root.in_lista(self.list)
            
    #def __repr__(self):
     #   if root is not None:
      #      self.lista_rep = root.lista_sinistra(self.list)
       #     return "Lista valori albero: {}".format(self.lista_rep)
            
    def __next__(self):
        if self.list == []:
            raise StopIteration("la lista {} è vuota!".format(self.list))
        
        else:
            current_node = self.list.pop()
            
            if current_node.destra is not None:
                current_node.destra.in_lista(self.list)
            return current_node.valore
            
```
