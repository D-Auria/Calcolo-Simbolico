```python
class B_Node:
    
    def __init__(self, val):
        self.value = val
        self.left = None   # "None are tree leaves"
        self.right = None
        
        
    def __repr__(self):
        
        x = "no branch"

        if self.left is None and self.right is None:
            return '(node {}, ({}))'.format( self.value, x)
        
        elif self.left is not None and self.right is None :
            return '(node {}, (<-- {},--> {}))'.format( self.value, self.left ,x)
        
        elif self.left is  None and self.right is not None :
            return '(node {}, (<-- {},--> {}))'.format(self.value, x, self.right)
        else :
            return '(node {}, (<-- {},--> {}))'.format( self.value, self.left, self.right)
        
        
    def insert_value(self, new_val):
        #print(new_val)
        if new_val < self.value:
            if self.left is None:
                self.left = B_Node(new_val)
            else:
                self.left.insert_value(new_val)
                
        elif new_val > self.value:
            if self.right is None:
                self.right = B_Node(new_val)
            else:
                self.right.insert_value(new_val)
        else:
            print("value equals node")
    
    
    def search(self, obj):
        
        if obj == self.value:
            return(True)
        
        elif obj < self.value:
            if self.left is None:
                return(False)
            else:
                return self.left.search(obj)
            
        elif obj > self.value:
            if self.right is None:
                return(False)
            else:
                return self.right.search(obj)
           
        
    def height_left(self):   
        
        h_sin = 0
         
        #if self is None:
         #   return 0
        if self.left is None:
            return 0
        else:
            if self.left is not None:
                #h_sin = self.left.height_left() + 1
                h_sin = self.left.height() + 1
                return h_sin
                


    def height_right(self):    
        
        h_des = 0
         
        #if self is None:
            #return 0
        if self.right is None:
            return 0
        else:
            if self.right is not None:
                #h_des = self.right.height_right() + 1
                h_des = self.right.height() + 1
                return h_des
                
        
    def height(self):    
        
        h_sin = self.height_left()
        h_des = self.height_right()

        if h_des> h_sin:  #I look for which of the branches of right or left is longer
            return h_des

        else:
            return h_sin
        
    def unbalance(self):
        sb = self.right.height_left() - self.height_right()
        return sb
    
    def min_node(self):
        if self.left is None:

            return self.value
        else:
            return self.left.min_node()
        
    def max_node(self):
        if self.right is None:
            return self.value
        else:
            return self.right.max_node()
        
    def estract_value(self):
        if self.left is None:
            return self.value
        elif self.right is None:
            return self.value
        
   
    def in_list(self, list):
        list.append(self)                 #I put the node in the list
        if self.left is not None:
            self.left.in_list(list)
        else:
            return list
 
```

```python
class BinaryTree:
    def __init__(self):
        self.root_node = None
        
    def __repr__(self):
        if self.root_node is None:
            return 'tree vuoto'
        else:
            print("tree: ")
            return self.root_node.__repr__()
    
    def InsertValue(self, value):
        if self.root_node is None:
            self.root_node = B_Node(value) 
            #print("ciao")
        else:
            self.root_node.insert_value(value) #be careful, here I'm using the insert_value attribute of B_Node
            
    def __contains__(self, value):
        if self.root_node is None:
            return(False)
        else:
            return self.root_node.search(value)
        
    def height_left_tree(self):
        if self.root_node is None:
            return -1
        else:
            return "left height of the tree: {}".format(self.root_node.height_left())
        
    def height_right_tree(self):
        if self.root_node is None:
            return -1
        else:
            return "right height of the tree: {}".format(self.root_node.height_right())
        
    def height_tree(self):
        if self.root_node is None:
            a = -1
            return "height of the tree: {}".format(a)  #by definition, height of an empty tree, without nodes is -1
        else:
            return "height of the tree: {}".format(self.root_node.height())
    
    def min_tree(self):
        if self.root_node is None:
            return "no value"
        else:
            return "minimum value of the tree: {}".format(self.root_node.min_node())
        
    def max_tree(self):
        if self.root_node is None:
            return "nessun value"
        else:
            return "max value of the tree: {}".format(self.root_node.max_node())
        
    def unbalance_tree(self):
        return "Unbalancement tree (left - right): {}".format(self.root_node.unbalance())
    
    #def lista_left_tree(self, list_tree):
     #   if self.root_node is None:
      #      return []
       # else:
        #    return self.root_node.lista_left(list_tree)
        
    def __iter__(self):
        return BSTIterator(self)
    
```

Of course, if we sort the array of random numbers before entering its values into the tree, we will have a tree that is completely unbalanced in one of the two branches and the tree "degenerates" into a list of right- or left-handed only nodes. If we leave the random array unordered (as in the first example), we will have an "average unbalanced" tree.

```python
class BSTIterator:
    
    def __init__(self, tree):
        self.tree = tree
        self.list = []
        root = self.tree.root_node      
        
        if root is not None:
            root.in_list(self.list)
            
    #def __repr__(self):
     #   if root is not None:
      #      self.lista_rep = root.list_left(self.list)
       #     return "Values List tree: {}".format(self.list_rep)
            
    def __next__(self):
        if self.list == []:
            raise StopIteration("list{} is empty!".format(self.list))
        
        else:
            current_node = self.list.pop()
            
            if current_node.right is not None:
                current_node.right.in_list(self.list)
            return current_node.value
            
```
